```python
filter = ['Nombre','Código NIF','Forma jurídica','Fecha constitución','Estado','Presidente','Director Ejecutivo','Participada - Nombre','Participada - Número de identificación BvD','Participada - Código ISO País']
renombre = ['nombre','CIF','forma','fconst','Estado','Pres','DE','participadaNombre','participadaId','participadaCodPais']
df1 = df.loc[:, filter]
df1.columns = renombre
df2=df1.dropna(how='all')
df2.head(20)
```

Leemos fila a fila del DF para crear las relaciones de las empresas participadas. Mientras el nombre de la emresa no cambie, el registro de n+1 se supone que es una relacion con la fila inicial principal

```python
dfrelaciones = pd.DataFrame(columns=('nombre', 'CIF', 'tipo','participadaNombre','participadaId','participadaCodPais'))
dft = df2[1:20]
dft2 =dft.reset_index()
dft2.head(20)
```

```python

for index, row in dft2.iterrows():
    print row['nombre']
    name = row['nombre']
    lastname = ''
    lastcif=''
    if not(pd.isnull(name)):
        lastname=name
        print lastname
        lastcif=row['CIF']
    if pd.isnull(name):
        dfrelaciones.loc[len(dfrelaciones)] = [lastname,lastcif,'participada',row['participadaNombre'],row['participadaId'],row['participadaCodPais']]
```
